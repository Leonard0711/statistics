## Pruebas Paramétricas
### Prueba t para 1 muestra

Prueba de hipótesis de 2 colas (bilateral)

**Problema**: Determinar si la media de los puntajes es diferente de 0.067.
* **Hipótesis nula (H0)**: La media de los puntajes no es diferente a 0.067.
* **Hipótesis alternativa (H1)**: La media de los puntajes es diferente a 0.067.

```python
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# generando los datos 
np.random.seed(1)
puntajes = np.random.normal(loc=0, scale=1, size=1000)
media_hipotetica = 0.067

t_stat, p_value = stats.ttest_1samp(puntajes, media_hipotetica)

print(f'El valor de t es: {np.round(t_stat, 5)}')
print('='*50)
print(f'El valor p es: {np.round(p_value, 5)}')
print('='*50)

alpha = 0.05
if p_value < alpha:
    print('Se rechaza la hipótesis nula')
else:
    print('No se rechaza la hipótesis nula')

df = 999 # grado de libertad
t_critico = stats.t.ppf(1 - alpha/2, df)
print('='*50)
print(f"Valor crítico t: ±{round(t_critico, 4)}")

# Rango de valores t para graficar
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10,6))
sns.kdeplot(puntajes)
plt.axvline(t_stat, color="red", linestyle="--", label=f"estadístico:{t_stat:.2f}")
plt.axvline(t_critico, color="green", linestyle="--", label=f"t_crítico: ±{t_critico:.2f}")
plt.axvline(-t_critico, color="green", linestyle="--")
plt.axvline(media_hipotetica, color="blue", linestyle="--", label=f"media_hipotetica: {media_hipotetica:.2f}")
plt.fill_between(x, y, where=(x > t_critico) | (x < -t_critico), color='orange', alpha=0.5, label='Región crítica')
plt.title('Prueba t de una muestra')
plt.xlabel('t')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()
```

Como el valor estadístico t cae fuera de la región crítica no se rechaza la hipótesis nula.

Prueba de hipótesis unilaterial (inferior)

**Problema:** Determinar si la media de los puntajes es menor a 0.067.
* **Hipótesis nula (H0):** La media de los puntajes no es menor o igual a 0.067.
* **Hipótesis alternativa (H1):** La media de los puntajes es menor a 0.067.

```python
np.random.seed(1)
puntajes = np.random.normal(loc=0, scale=1, size=1000)
media_hipotetica = 0.067
t_stat, p_value = stats.ttest_1samp(puntajes, media_hipotetica,alternative='less')

print(f'El valor de t es: {np.round(t_stat, 5)}')
print('='*50)
print(f'El valor p es: {np.round(p_value, 5)}')
print('='*50)

alpha = 0.05
if p_value < alpha:
    print('Se rechaza la hipótesis nula')
else:
    print('No se rechaza la hipótesis nula')

df = 999 # grado de libertad
t_critico = stats.t.ppf(alpha, df)
print('='*50)
print(f"Valor crítico t: {round(t_critico, 4)}")

# Rango de valores t para graficar
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10,6))
sns.kdeplot(puntajes)
plt.axvline(t_stat, color="red", linestyle="--", label=f"estadístico:{t_stat:.2f}")
plt.axvline(t_critico, color="green", linestyle="--", label=f"t_crítico: {t_critico:.2f}")
# plt.axvline(-t_critico, color="green", linestyle="--")
plt.axvline(media_hipotetica, color="blue", linestyle="--", label=f"media_hipotetica: {media_hipotetica:.2f}")
plt.fill_between(x, y, where=x < t_critico, color='orange', alpha=0.5, label='Región crítica (α = 0.05)')
plt.title('Prueba t de una muestra (una cola inferior)')
plt.xlabel('t')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()
```

Como el valor estadístico t cae fuera de la región crítica no se rechaza la hipótesis nula.

Prueba de hipótesis unilateral (superior)

**Problema:** Determinar si la media de los puntajes es mayor a 0.067.
* Hipótesis nula (H0): La media de los puntajes no es igual o mayor a 0.067.
* Hipótesis alternativa (H1): La media de los puntajes es mayor a 0.067.

```python
import numpy as np
from scipy import stats

np.random.seed(1)
puntajes = np.random.normal(loc=0, scale=1, size=1000)
media_hipotetica = 0.067

t_stat, p_value = stats.ttest_1samp(puntajes, media_hipotetica, alternative='greater')
print(f'El valor de t es: {np.round(t_stat, 5)}')
print('='*50)
print(f'El valor p es: {np.round(p_value, 5)}')
print('='*50)
alpha = 0.05
if p_value < alpha:
    print('Se rechaza la hipótesis nula')
else:
    print('No se rechaza la hipótesis nula')

df = 999 # grado de libertad
t_critico = stats.t.ppf(1 - alpha, df)
print('='*50)
print(f"Valor crítico t: {round(t_critico, 4)}")

# Rango de valores t para graficar
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10,6))
sns.kdeplot(puntajes)
plt.axvline(t_stat, color="red", linestyle="--", label=f"estadístico:{t_stat:.2f}")
plt.axvline(t_critico, color="green", linestyle="--", label=f"t_crítico: {t_critico:.2f}")
plt.axvline(media_hipotetica, color="blue", linestyle="--", label=f"media_hipotetica: {media_hipotetica:.2f}")
plt.fill_between(x, y, where=x > t_critico, color='orange', alpha=0.5, label='Región crítica (α = 0.05)')
plt.title('Prueba t de una muestra (una cola superior)')
plt.xlabel('t')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.show()
```

Como el valor estadístico t cae fuera de la región crítica no se rechaza la hipótesis nula.

### Prueba t para 2 muestras independientes

Prueba de hipótesis de 2 colas (bilateral)

**Problema:** Determinar si hay una diferencia entre la media de los puntajes del grupo A y la media
de los puntajes del grupo B.
* Hipótesis nula (H0): La media de los puntajes del grupo A y la media de los puntajes del
grupo B no son diferentes.
* Hipótesis alternativa (H1): La media de los puntajes del grupo A y la media de los
puntajes del grupo B son diferentes.

```python
import numpy as np
from scipy import stats

np.random.seed(2)
grupo_a = np.random.normal(loc=0, scale=1, size=100)
grupo_b = np.random.normal(loc=0, scale=1, size=100)
t_stat, p_value = stats.ttest_ind(grupo_a, grupo_b)

print(f'El valor de t es: {np.round(t_stat, 5)}')
print('='*50)
print(f'El valor p es: {np.round(p_value, 5)}')
print('='*50)

alpha = 0.05
if p_value < alpha:
    print('Se rechaza la hipótesis nula')
else:
    print('No se rechaza la hipótesis nula')

df = len(grupo_a) + len(grupo_b) - 2 # grado de libertad
t_critico = stats.t.ppf(1 - alpha/2, df)
print('='*50)
print(f"Valor crítico t: {round(t_critico, 4)}")

# Rango de valores t para graficar
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue')
plt.axvline(t_stat, color="red", linestyle="--", linewidth=2, label=f"estadístico = {t_stat:.2f}")
plt.axvline(t_critico, color="green", linestyle="--" , label=f"t crítico: ±{t_critico:.2f}")
plt.axvline(-t_critico, color="green", linestyle="--")
plt.fill_between(x, y, where=(x > t_critico) | (x < -t_critico), color='orange', alpha=0.5, label='Región crítica')
plt.title('Distribución t para prueba de dos muestras independientes (dos colas)')
plt.xlabel('Valores t')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.tight_layout()
plt.show()
```

Como el valor estadístico t cae fuera de la región crítica no se rechaza la hipótesis nula.

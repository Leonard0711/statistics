```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt 

numeros = np.random.randint(1, 100, 80)
conteo_1 = Counter(numeros)
# Al agrupar nuevamente en Counter permite seguir usando métodos específicos como:
# most_common(n): devuelve los n elementos más frecuentes.
# elements(): devuelve un iterador sobre los elementos, repitiendo cada uno según su frecuencia.
# Operaciones entre Counter como suma, resta, intersección, etc.
conteo = Counter({int(k):value for k, value in conteo_1.items()})
xs = range(101)
ys = [conteo[x] for x in xs]

plt.bar(xs, ys)
plt.show()
```
